* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2)

left, bottom, width, height = [0, 0, 1, 1]
ax1 = fig.add_axes([left, bottom, width, height])


left, bottom, width, height = [0.2, 0.5, 0.2, 0.2]
ax2 = fig.add_axes([left, bottom, width, height])

plt.subplots_adjust(wspace=0.3)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

ax1.set_xlim([0, 1])
ax1.set_ylim([0, 1])

ax2.set_xlim([0, 1])
ax2.set_ylim([0, 1])

ax1.plot([0, 1], [0, 1])
ax2.plot([0.2, 0.8], [0.5, 0.5])

plt.show()
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
x = np.linspace(-10, 10000, 100)
y = 2 * x

plt.plot(x, y, linestyle='--', linewidth=2, color='darkblue', label='y = 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = 2x')
plt.legend()

plt.xlim(0, 100)
plt.ylim(0, 200)

plt.show()


x1 = np.linspace(-10, 100, 100)
y1 = x1**2

plt.plot(x1, y1, linestyle='-', linewidth=2, color='red', label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2')
plt.legend()

plt.xlim(0, 100)
plt.ylim(0, 10000)

plt.show()

```
