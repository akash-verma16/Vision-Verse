```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_sampling(image, sampling_factor):
    # Using OpenCV resize function for downsampling
    sampled_image = cv2.resize(image, (image.shape[1] // sampling_factor, image.shape[0] // sampling_factor))
    return sampled_image

# Load an example image
original_image = cv2.imread('/content/drive/MyDrive/Lenna_(test_image).png')

# Display the original image
plt.subplot(131)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Apply sampling with different factors
sampling_factors = [2, 4]
for i, factor in enumerate(sampling_factors):
    sampled_image = image_sampling(original_image, factor)
    plt.subplot(1, 3, i + 2)
    plt.title(f'Sampled {factor}x')
    plt.imshow(cv2.cvtColor(sampled_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

plt.show()
```

**2.Quantisation**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_quantization(image, levels):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Perform quantization using numpy
    quantized_image = np.floor(gray_image / (256 / levels)) * (256 / levels)

    # Convert back to 3-channel image
    quantized_image = cv2.cvtColor(quantized_image.astype(np.uint8), cv2.COLOR_GRAY2BGR)

    return quantized_image

# Load an example image
original_image = cv2.imread('/content/drive/MyDrive/Lenna_(test_image).png')

# Display the original image
plt.subplot(131)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Apply quantization with different levels
quantization_levels = [4, 8, 16]
for i, levels in enumerate(quantization_levels):
    plt.subplot(1, len(quantization_levels), i + 1)  # Adjusted the third argument
    plt.title(f'Quantized {levels} levels')
    plt.imshow(cv2.cvtColor(quantized_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')


plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_sampling(input_image, sampling_factor):
   # Perform image sampling using OpenCV's resize function with interpolation
    downsampled_image = cv2.resize(input_image,
                                   None,
                                   fx=1/sampling_factor,
                                   fy=1/sampling_factor,
                                   interpolation=cv2.INTER_NEAREST)

    return downsampled_image

# Load the image (replace 'path_to_your_image' with the actual path to your image)
original_image = cv2.imread('/content/drive/MyDrive/Lenna_(test_image).png')

# Display the original image
plt.figure(figsize=(8, 8))
plt.subplot(331)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Apply sampling for different factors and display the downsampled images
sampling_factors = [2, 4, 8, 16, 32, 64]

for i, factor in enumerate(sampling_factors):
    downsampled_image = image_sampling(original_image, factor)
    plt.subplot(3, 3, i + 2)
    plt.imshow(cv2.cvtColor(downsampled_image, cv2.COLOR_BGR2RGB))
    plt.title(f'Sampled 1/{factor}x')

plt.tight_layout()
plt.show()
```

```python
def image_quantization(input_image, quantization_levels):
    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)

    # Perform image quantization
    quantized_image = np.floor_divide(grayscale_image, 256 // quantization_levels) * (256 // quantization_levels)

    return quantized_image

# Display the original image
plt.figure(figsize=(8, 8))
plt.subplot(331)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Apply quantization for different levels and display the quantized images
quantization_levels = [4, 8, 16, 32, 64, 128, 256]

for i, levels in enumerate(quantization_levels):
    quantized_image = image_quantization(original_image, levels)
    plt.subplot(3, 3, i + 2)
    plt.imshow(quantized_image, cmap='gray')
    plt.title(f'Quantized to {levels} levels')

plt.tight_layout()
plt.show()
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_sampling(image, sampling_factor, interpolation=cv2.INTER_NEAREST):
    # Use OpenCV resize function for downsampling with specified interpolation method
    sampled_image = cv2.resize(image, (image.shape[1] // sampling_factor, image.shape[0] // sampling_factor),
                               interpolation=interpolation)
    return sampled_image



# Display the original image
plt.subplot(131)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Apply sampling with different factors using nearest-neighbor interpolation
sampling_factors = [2, 4,8,16,32,64,128]
for i, factor in enumerate(sampling_factors):
    sampled_image = image_sampling(original_image, factor, interpolation=cv2.INTER_NEAREST)
    plt.subplot(1, 3, i + 2)
   # plt.title(f'Sampled {factor}x (Nearest Neighbor)')
    plt.imshow(cv2.cvtColor(sampled_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

plt.show()
```
