Hough Transform


```python
from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

#Read Image
image1=cv2.imread(filename,cv2.IMREAD_COLOR)

#Convert to gray
image_gray=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

#Gaussian Blur
image_blur=cv2.GaussianBlur(src=image_gray, ksize=(3,3),sigmaX=0,sigmaY=0)

#canny Edge
image_canny=cv2.Canny(image=image_blur,threshold1=5,threshold2=130)

#Hough Transform
img_hough=cv2.HoughLinesP(image_canny,1,np.pi/180, threshold=60,minLineLength=10,maxLineGap=250)

for line in img_hough:
   x1,y1,x2,y2= line[0]
   cv2.line(image1,(x1,y1),(x2,y2),(255,0,0),4)

#Show

cv2_imshow(img_hough)
cv2_imshow(image1)
```

Laplacian Edge detection

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload an image file
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

    # Read the uploaded image
    image = cv2.imdecode(np.frombuffer(uploaded[filename], dtype=np.uint8), cv2.IMREAD_GRAYSCALE)

    # Apply Gaussian Blur
    image1 = cv2.GaussianBlur(src=image, ksize=(3, 3), sigmaX=0, sigmaY=0)

    # Apply Laplacian edge detection
    laplacian = cv2.Laplacian(image1, cv2.CV_64F)

    # Convert the result to uint8
    laplacian = np.uint8(np.absolute(laplacian))

    # Display the original and Laplacian images
    plt.figure(figsize=(10, 5))

    plt.subplot(121), plt.imshow(image, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(122), plt.imshow(laplacian, cmap='gray')
    plt.title('Laplacian Edge Detection'), plt.xticks([]), plt.yticks([])

    plt.show()
```

 Edge Detection Using Gradients-Sobel,Prewitt,Robert Operators

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload an image file
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'Uploaded file: "{filename}" with {len(uploaded[filename])} bytes')

    # Read the uploaded image
    image = cv2.imdecode(np.frombuffer(uploaded[filename], dtype=np.uint8), cv2.IMREAD_GRAYSCALE)

    # Sobel Operator
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    sobel_edge = np.sqrt(sobel_x**2 + sobel_y**2)

    # Prewitt Operator
    prewitt_kernel_x = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
    prewitt_kernel_y = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    prewitt_x = cv2.filter2D(image, cv2.CV_64F, prewitt_kernel_x)
    prewitt_y = cv2.filter2D(image, cv2.CV_64F, prewitt_kernel_y)
    prewitt_edge = np.sqrt(prewitt_x**2 + prewitt_y**2)

    # Roberts Operator
    roberts_kernel_x = np.array([[1, 0], [0, -1]])
    roberts_kernel_y = np.array([[0, 1], [-1, 0]])
    roberts_x = cv2.filter2D(image, cv2.CV_64F, roberts_kernel_x)
    roberts_y = cv2.filter2D(image, cv2.CV_64F, roberts_kernel_y)
    roberts_edge = np.sqrt(roberts_x**2 + roberts_y**2)

    # Display the results
    plt.figure(figsize=(15, 5))

    plt.subplot(131), plt.imshow(sobel_edge, cmap='gray')
    plt.title('Sobel Edge Detection'), plt.xticks([]), plt.yticks([])

    plt.subplot(132), plt.imshow(prewitt_edge, cmap='gray')
    plt.title('Prewitt Edge Detection'), plt.xticks([]), plt.yticks([])

    plt.subplot(133), plt.imshow(roberts_edge, cmap='gray')
    plt.title('Roberts Edge Detection'), plt.xticks([]), plt.yticks([])

    plt.show()
```
